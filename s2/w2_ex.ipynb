```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

Load using Pandas

```python
df1 = pd.read_csv('data1.tsv', sep="\t") 
df2 = pd.read_csv('data2.tsv', sep="\t") 
df3 = pd.read_csv('data3.tsv', sep="\t") 
df4 = pd.read_csv('data4.tsv', sep="\t") 
```

Load Using Numpy

```python
d1 = np.loadtxt('data1.tsv')
d2 = np.loadtxt('data2.tsv')
d3 = np.loadtxt('data3.tsv')
d4 = np.loadtxt('data4.tsv')

mu_d1 = d1.mean(axis=0)
mu_d2 = d2.mean(axis=0)
mu_d3 = d3.mean(axis=0)
mu_d4 = d4.mean(axis=0)

print(f'X Mean dataFrame 1 {mu_d1[0]:.2f}')
print(f'Y Mean dataFrame 1 {mu_d1[1]:.2f}')
print(f'X Mean dataFrame 2 {mu_d2[0]:.2f}')
print(f'Y Mean dataFrame 2 {mu_d2[1]:.2f}')
print(f'X Mean dataFrame 3 {mu_d3[0]:.2f}')
print(f'Y Mean dataFrame 3 {mu_d3[1]:.2f}')
print(f'X Mean dataFrame 4 {mu_d4[0]:.2f}')
print(f'Y Mean dataFrame 4 {mu_d4[1]:.2f}')
```

Variance of all datadets

```python
DF = [d1, d2, d3, d4]

for i in range(4):
    print(f'The variance on X axis for dataSet {i} is {np.var(DF[i][:,0]):.03f}')
    print(f'The variance on Y axis for dataSet {i} is {np.var(DF[i][:,1]):.03f}')
```

Parson Correlation 
As it's going to be a Symetric matrix with ones in the diagonal, to get the correlation between x and y e'll have to get the correlation of the off diagonals, in this case they should be the same.

```python
for i in range(4):
    corr = np.corrcoef(DF[i][:,0], DF[i][:,1])[0,1]
    print(f'Parson Correlation for DatasSet {i} is: {corr:.03f}')
```

Linear Regresion to fit a straigth line.
f(x) = ax + b
Where a is the slope and b is the intercept value>

```python
slopes = []
intercepts = []
r_values = []
p_values = []
std_errors = []

for i in range(4):
    slope, intercept, r_value, p_value, std_err = linregress(DF[i][:,0], DF[i][:,1])
    slopes.append(slope)
    intercepts.append(intercept)
    r_values.append(r_value)
    p_values.append(p_value)
    std_errors.append(std_err)
    print(f'Data Set {i}:\n\tSlope (a) = {slope:.02f}\n\tIntercept (b) = {intercept:.02f}')

```

We can see that all the data sets have a high corrleation between x and y axis. Specifically, the first 3 data Sets have the same correlation between both axis. The fourth dataset has a higher value, 1 decimal. 
At first glance that could seem strange, but if we Calculate the Line equation, we can see that both Slopes and Intercept values have same values. That indicates that probably all datasets are linear combination, not independent.

```python
fig, ax = plt.subplots(2,2, figsize=(15,15))

ax[0,0].scatter(d1[:,0], d1[:,1])
ax[0,0].plot(d1[:,0], slopes[0]*d1[:,0]+intercepts[0], '--r')
ax[0,0].set_title('Data Set 1')  
ax[0,0].set_xlim([2,19])

ax[0,1].scatter(d2[:,0], d2[:,1])
ax[0,1].plot(d2[:,0], slopes[1]*d2[:,0]+intercepts[1], '--r')
ax[0,1].set_title('Data Set 2') 
ax[0,1].set_xlim([2,19])

ax[1,0].scatter(d3[:,0], d3[:,1])
ax[1,0].plot(d3[:,0], slopes[2]*d3[:,0]+intercepts[2], '--r')
ax[1,0].set_title('Data Set 3') 
ax[1,0].set_xlim([2,19])

ax[1,1].scatter(d4[:,0], d4[:,1])
ax[1,1].plot(d4[:,0], slopes[3]*d4[:,0]+intercepts[3], '--r')
ax[1,1].set_title('Data Set 4') 
ax[1,1].set_xlim([2,19])
```

The Idea of that exercise is that even if the data is really different, as we can see in the example above, we can have same variances, means, slopes and intercepts... 
We can see in the above exaple that the data sets are quite differents, in fact Dataset 4 has almost all the points in the x point of value 8, and even then we get the same results. 
The thing is that statistics could lead to incorrect assumptions of data if we don't approach correctly the problem. We should base our assumptions in statistics, data visualitzation and data knwoledge. 

**Data vs MetaData**

While data can simply be a piece of information, a list of measurements, or observations, a story or a description of a certain entity, metadata specifies information about the original data which assists in identifying the nature and features of that data.
We can infer information from pictures to render the map 


**Human Eye**

Humans do have Statistical intuition but sometimes could be hard to see differences when talking about numbers, specifically because we don't get the full perspective of what we are working on. as he said in the example, we could percieve a better difference in the higth of a building for example if we are in front of a buiding of 3m high or one of 100m. 


**Simpson's Paradox**

Simpson's Paradox is a statistical phenomenon where an association between two variables in a population emerges, disappears or reverses when the population is divided into subpopulations.


**Exploratory and Explanatory:**

Exploratory analysis is basically exploring the data, get the insights of what's the data explaining and how the data is behaving. Expanatory analyis is done once you have made the exploratory analysis. You try to communicate with results and plots the key features of the data.
```python
df_tmp = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")
```

```python
df_tmp['Date'] = pd.to_datetime(df_tmp['Date'], format="%m/%d/%Y")
df_tmp['Hour'] = pd.to_datetime(df_tmp['Time'].astype(str)).dt.hour
df_tmp['Time'] = pd.to_datetime(df_tmp['Time'], format= '%H:%M').dt.time
```

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
len(focuscrimes)
```

```python
df_focus = df_tmp[df_tmp.Category.isin(focuscrimes)]
len(df_focus.columns)
```

```python
grouped = df_focus.groupby(['Category', df_focus.DayOfWeek]).count()['PdId']
```

Daily Plots

```python
ax = grouped.unstack(level=0).plot(kind='bar', subplots=True, rot=90, figsize=(12, 20), layout=(7, 2))
plt.tight_layout()
```

```python
grouped = df_focus.groupby(['Category', df_focus.Date.dt.month]).count()['PdId']
ax = grouped.unstack(level=0).plot(kind='bar', subplots=True, rot=90, figsize=(12, 20), layout=(7, 2))
plt.tight_layout()
```

Hourly plots

```python
grouped = df_focus.groupby(['Category', df_focus.Hour]).count()['PdId']
ax = grouped.unstack(level=0).plot(kind='bar', subplots=True, rot=90, figsize=(12, 20), layout=(7, 2))
plt.tight_layout()
```

```python
df_focus['TimeExt'] = df_focus['Date'].dt.weekday*df_focus['Hour']
```

```python
grouped = df_focus.groupby(['Category', df_focus.TimeExt]).count()['PdId']
ax = grouped.unstack(level=0).plot(kind='bar', subplots=True, rot=90, figsize=(12, 20), layout=(7, 2))
plt.tight_layout()
```
