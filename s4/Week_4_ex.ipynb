-Area: If we usea area as an encoder it will be hard to see the diferences, because some tinny changes in area that might be significant to the result might not be significant to the eye. The following video perfectly shows how tinny changes in area are not percieved by human eye.

```python
df = pd.read_csv("../data/Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv", index_col=False)
```

```python
cat = 'DRUG/NARCOTIC'
df_drug = df[df.Category==cat][['PdId', 'X', 'Y', 'Date', 'Time']]
df_drug['Date'] = pd.to_datetime(df_drug['Date'], format="%m/%d/%Y")
start = datetime.strptime('01/01/2006', "%m/%d/%Y")
end = datetime.strptime('02/01/2006', "%m/%d/%Y")
df_drug_def = df_drug[(df_drug.Date>start)&(df_drug.Date<end)&(df_drug.Time>'13:00')&(df_drug.Time<='14:00')].sort_values('Time')
```

```python
SF_map_DRUG = folium.Map([37.7749, -122.4194], zoom_start=13)

for i in df_drug_def.iterrows():
    lon = i[1]['X']
    lat = i[1]['Y']
    folium.Marker([lat, lon]).add_to(SF_map_DRUG)

SF_map_DRUG
```

```python
sex_cat = 'SEX OFFENSES, NON FORCIBLE'
df_sex = df[df.Category == sex_cat]
```

```python
SF_map_sex = folium.Map([37.7749, -122.4194], zoom_start=13)
```

```python
df_sex['X'] = df_sex['X'].astype(float)
df_sex['Y'] = df_sex['Y'].astype(float)
```

```python
heat_df = df_sex[['X','Y']]
heat_data = [[row['Y'],row['X']] for index, row in heat_df.iterrows()]
```

```python
HeatMap(heat_data).add_to(SF_map_sex)
SF_map_sex
```

```python
SF_map_sex2 = folium.Map([37.7749, -122.4194], zoom_start=13)
HeatMap(heat_data, min_opacity=10, radius=10, blur=15).add_to(SF_map_sex2)
SF_map_sex2
```

The radius and the blur are argumens that vary the radius of the circles that are displayed on the map, the heatmap circles. The blur specifyies the intensity of the bluring of the heatmap circles.
